```python
'''# üß† Day 1 ‚Äî PyTorch Tensors, Datasets & DataLoader

---

## 1Ô∏è‚É£ Introduction

On Day 1, I began my MLOps learning journey by diving into **PyTorch fundamentals**.  
I learned about:
- Tensors (PyTorch‚Äôs version of NumPy arrays)
- Basic tensor operations (reshape, slicing, broadcasting)
- How to organize data using `TensorDataset` and `DataLoader`
- The concept of **mini-batching** and **shuffling**

These concepts form the **foundation of all model training** in PyTorch ‚Äî they make it possible to efficiently handle data, feed it to models, and optimize performance during training.

---

## 2Ô∏è‚É£ Tensor Operations

Here, I created and experimented with PyTorch tensors ‚Äî exploring shapes, reshaping, slicing, and broadcasting.

'''
import torch

# Create tensors
x = torch.arange(10)
print("Original tensor:", x)

# Reshape to 2D
x = x.reshape(5, 2)
print("\nReshaped tensor:\n", x)

# Basic operations
y = torch.ones((5, 2))
print("\nAddition:\n", x + y)
print("\nMultiplication:\n", x * 2)

# Slicing
print("\nSliced (first two rows):\n", x[:2])

# Broadcasting example
a = torch.tensor([[1], [2], [3]])
b = torch.tensor([10, 20, 30])
print("\nBroadcasted sum:\n", a + b)



```

```python
from torch.utils.data import TensorDataset, DataLoader

# Create input (X) and output (y)
X = torch.arange(10).float().reshape(-1, 1)
y = 2 * X + 1

# Combine into dataset
dataset = TensorDataset(X, y)

# Create DataLoader for batching
loader = DataLoader(dataset, batch_size=2, shuffle=True)

# Iterate through a few mini-batches
for xb, yb in loader:
    print(f"X batch: {xb.squeeze().tolist()} ‚Üí y batch: {yb.squeeze().tolist()}")
    break
```

'''TensorDataset groups features & labels together

DataLoader splits data into mini-batches (size = 2 here)

shuffle=True randomizes data each epoch ‚Äî important to prevent model overfitting to data order'''

```python
import matplotlib.pyplot as plt

plt.scatter(X, y, color='blue', label='Data Points')
plt.title("Linear Relationship: y = 2x + 1")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
